```python
from sim_launch_py.external.molecules import findTorsionalAngles
from sim_launch_py.external.gaussianFit import gaussianFit as gaussfit
import matplotlib.pyplot as plt

dihedrals=findTorsionalAngles('dih/ccdc-xxxii.pdb')
```

```python
print("There are {} dihedrals:\n".format(len(dihedrals)))

print(dihedrals)
```

```python
colvar='dih/COLVAR'

angles=[[] for d in dihedrals]
time=[]

commentchars=['@','#']

with open(colvar,'r') as f:
    for line in f:
        if line[0] not in commentchars:
            cols=line.split()
            time.append(float(cols[0]))
            for icol in range(1,len(cols)):
                angles[icol-1].append(float(cols[icol]))
```

```python
fig,ax=plt.subplots(nrows=1,ncols=1)

datacol=0

ax.scatter(time,angles[datacol],s=0.05)
ax.set_ylim([-3.14,3.14])
plt.show

centers,sigma=gaussfit(colvar,datacol=datacol+1,plotFit=True)
```
